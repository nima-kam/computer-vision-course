# Panorama with OpenCV

```python
import numpy as np
import cv2
import glob
import matplotlib.pylab as plt
```

```python
# read input victoria images and show them in a row together

images_path = glob.glob('images/victoria*.png')
list_imgs = []
for i in range(len(images_path)):

    image = cv2.imread(str(images_path[i]))  
    # image = cv2.resize(image,(600,800))
    # print("i: ",images_path[i],"\nimg:",image.shape)
    list_imgs.append(image)

width = 2
height = (len(list_imgs)+ width-1)//width
fig=plt.figure()
for i in range(len(list_imgs)):
    fig.add_subplot(width,height,i+1)
    plt.imshow(list_imgs[i])
    plt.axis('off')
    plt.title(f"Victoria{i+1}")
```

```python
# initialize OpenCV's image sticher object and then perform the image stitching on input images
# reference: https://docs.opencv.org/4.x/d2/d8d/classcv_1_1Stitcher.html#a114713924ec05a0309f4df7e918c0324ac706a6a118a4a648fef8b2fca5950e2a

stitchy=cv2.Stitcher.create(1)

(dummy,output)=stitchy.stitch(list_imgs)

 
```

```python
# show victoria panorama

if dummy != cv2.STITCHER_OK: # check if image created successfully
  
    print("stitching ain't successful. ",dummy)
else:
    print('Panorama is ready!!!')
    plt.figure()
    plt.imshow(output)
    plt.axis('off')
    plt.title(f"Victoria panorama")
```
