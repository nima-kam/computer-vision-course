```python
# Import required packages
import cv2
import pytesseract
from google.colab.patches import cv2_imshow
from PIL import Image
import matplotlib.pyplot as plt
```

```python

##################### your code ################################
img = cv2.imread("/content/drive/MyDrive/ai/computer vision/university/royan.jpg")
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

tresh,otsu=cv2.threshold(gray,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
adaptive=cv2.adaptiveThreshold(gray,250,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,3)
# print(otsu)
plt.figure()
plt.imshow(otsu,cmap='gray')
plt.figure()
plt.imshow(adaptive,cmap='gray')

kernel=cv2.getStructuringElement(
    cv2.MORPH_DILATE,(3,3)
)
result = cv2.dilate(adaptive,kernel)
# print(result," , kernel")

plt.figure()
plt.title('after dilation')
plt.imshow(result ,cmap='gray')

################################################################
# Finding contours
## result = result otso|adaptive algorithm

print(pytesseract.image_to_string( result , lang='fas'))
```

```python
# Import required packages
import cv2
import pytesseract
from google.colab.patches import cv2_imshow
from PIL import Image
##################### your code ################################
img = cv2.imread("/content/drive/MyDrive/ai/computer vision/university/mypic.png")
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

tresh,otsu=cv2.threshold(gray,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
adaptive=cv2.adaptiveThreshold(gray,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,3)
# print(otsu)
plt.figure()
plt.title('otsu threshold')
plt.imshow(otsu,cmap='gray')
plt.figure()
plt.title('adaptive threshold')
plt.imshow(adaptive,cmap='gray')

kernel=cv2.getStructuringElement(
    cv2.MORPH_DILATE,(3,3)
)
dilation = cv2.dilate(adaptive,kernel)
print(dilation," , kernel")

plt.figure()
plt.title('after dilation')
plt.imshow(dilation,cmap='gray')

################################################################
# Finding contours
im2 = img.copy()
contours, hierarchy = cv2.findContours(dilation, cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)

for cnt in contours:
  x, y, w, h = cv2.boundingRect(cnt)
	# Drawing a rectangle on copied image
  rect = cv2.rectangle(im2, (x, y), (x + w+4, y + h+5), (0, 255, 0), 2)
	# Cropping the text block for giving input to OCR
  cropped = im2[y:y + h, x:x + w]
  print(pytesseract.image_to_string(cropped, lang='fas'))

plt.figure()
plt.title('rect')
plt.imshow(im2,cmap='gray')
```

```python
image_path_in_colab='/content/drive/MyDrive/ai/computer vision/university/mypic.png'
extractedInformation = pytesseract.image_to_string(image_path_in_colab, lang='fas')
print(extractedInformation)
```
